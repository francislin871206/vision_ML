思考一下我們前面有提到圖片是矩陣，但維度可能會不一樣
例如灰階圖只有兩個維度，RGB 彩圖則有 3 個維度

假如今天我們把 RGB 3 個維度拆開來看會有甚麼不同的效果呢？

```python
import cv2
import numpy as np
img_pth = 'data/view.png'

img = cv2.imread(img_pth,cv2.IMREAD_COLOR)
```

```python
#正常版本
while True:
    cv2.imshow('bgr',img)
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()
        break
```

```python
def MergeThreeColor(b,g,r):
    img = cv2.merge([b,g,r])
    return img   
def SplitThreeColor(img):
    b,g,r = cv2.split(img)
    zeros = np.zeros(img.shape[:2],dtype = np.uint8)
    img_b = MergeThreeColor(b,zeros,zeros)
    img_g = MergeThreeColor(zeros,g,zeros)
    img_r = MergeThreeColor(zeros,zeros,r)
    print("which color do you want to see ?")
    color = input()
    if color == "blue":
        while True:
            cv2.imshow('blue',img_b)
            k = cv2.waitKey(0)
            if k == 27:
                cv2.destroyAllWindows()
                break
    if color == "green":
        while True:
            cv2.imshow('green',img_g)
            k = cv2.waitKey(0)
            if k == 27:
                cv2.destroyAllWindows()
                break
    if color == "red":
        while True:
            cv2.imshow('red',img_r)
            k = cv2.waitKey(0)
            if k == 27:
                cv2.destroyAllWindows()
                break
    

    
    
```

```python
SplitThreeColor(img)
```
