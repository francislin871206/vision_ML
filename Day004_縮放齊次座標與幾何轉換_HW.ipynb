```python
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt
def Img_Show_cv2(img,img_name):
    while True:
        cv2.imshow(img_name,img)
        k = cv2.waitKey(0)
        if k == 27:
            cv2.destroyAllWindows()
            break
# cv2 預設 color space 是 BGR，但 matplotlib，
# 顯示是以 RGB 來顯示，故需要做順序的調整
img_pth = 'data/lena.png'
img = cv2.imread(img_pth)
img_grb = img[...,[1,2,0]]
img_rgb = img[...,[2,1,0]]
plt.imshow(img_rgb)
plt.axis('off')
plt.show()

plt.imshow(img)
plt.axis('off')
plt.show()

plt.imshow(img_grb)
plt.axis('off')
plt.show()
""" Your Code Here: 讀取 lena.png 並使用 matplotlib 顯示 """
```

# 進行放大的比較

```python
# 方便顯示與進行縮放
interpolation_lst = [cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_AREA, cv2.INTER_CUBIC, cv2.INTER_LANCZOS4]
interpolation_name = ['INTER_LANCZOS4', 'INTER_LINEAR', 'INTER_AREA', 'INTER_CUBIC', 'INTER_LANCZOS4']

# 定義欲放大與縮小的尺寸
small_length, big_length = int(img.shape[0] / 4), int(4 * img.shape[0])
print(big_length)
print(img.shape[0])
# 定義作圖版面大小
plt.figure(figsize = (25, 5))

# 開始進行5種插值法的放大比較
for i in range(len(interpolation_lst)):
    start = time.time() # 按下碼表計時
    img_chg = cv2.resize(img_rgb,(big_length,big_length),interpolation_lst[i])
    """ Your Code Here: 插值演算法實作(Hint: cv2.resize) """
    end = time.time() # 結束計時
    # 開始做圖
    plt.subplot(1,5,i+1,title = "{} end in {}".format(interpolation_name[i],end - start))
    plt.imshow(img_chg)
    plt.axis('off')
    """ Your Code Here: 開始做圖 """
plt.show()
```

# 進行縮小的比較

```python
plt.figure(figsize = (25, 5))

# 開始進行5種插值法的縮小比較
for i in range(len(interpolation_lst)):
    start = time.time() # 按下碼表計時
    img_small = cv2.resize(img_rgb,(small_length,small_length),interpolation_lst[i])
    end = time.time() # 結束計時
    plt.subplot(1,5,i+1,title = "{} end in {}".format(interpolation_name[i],end - start))
    plt.imshow(img_small)
plt.show()
    
    # 開始做圖
```
