## 先讀取並顯示圖片

```python
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt

img = cv2.imread('data/lena.png')

# cv2 預設 color space 是 BGR，但 matplotlib，
# 顯示是以 RGB 來顯示，故需要做順序的調整
plt.imshow(img[:, :, [2, 1, 0]])
plt.axis('off')
plt.show()
```

## 1. 將 RGB 轉換為 HSV，分別印出 Hue, Saturation, Value

```python
# 將 RGB 轉換為 HSV
""" Your Code Here: 讀取 lena.png 並使用 matplotlib 顯示 """
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
plt.axis('off')

# 分別印出 Hue, Saturation, Value
title = ['Hue', 'Saturation', 'Value']
h,s,v = cv2.split(hsv)
plt.figure(figsize = (30,10))

plt.subplot(1,3,1,title = title[0])
plt.imshow(h)
plt.axis('off')

plt.subplot(1,3,2,title = title[1])
plt.imshow(s)
plt.axis('off')

plt.subplot(1,3,3,title = title[2])
plt.imshow(v)
plt.axis('off')

plt.show()


""" Your Code Here: 讀取 lena.png 並使用 matplotlib 顯示 """
```

## 2. 對 Value 進行 Histogram Equalization，並印出處理結果(處理完的圖片和處理完的 Value)

```python
# 先印出 HE 前的圖片與 value
plt.imshow(v)
plt.axis('off')
plt.show()
print(v)
# 印出 HE 後的圖片與 value

v_hist = cv2.equalizeHist(v)
plt.imshow(v_hist)
plt.axis('off')
plt.show()
print(v_hist)

```

## 3. 將 RGB 轉換為 Gray level，再進行單值二值化及自適應二值化

```python
# 將圖片轉換為灰階圖片

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

plt.imshow(gray,cmap = 'gray')
plt.axis('off')
plt.show()

# Single Binarization
ret,img_thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
print(ret)
plt.imshow(img_thresh,cmap = 'gray')
plt.axis('off')
plt.show()

# Adaptive Binarization
adap_bin = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,9,0)
print(adap_bin)
plt.imshow(adap_bin,cmap = 'gray')
plt.axis('off')
plt.show()
```
